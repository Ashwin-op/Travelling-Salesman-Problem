```python
import glob
import os

import pandas as pd
from natsort import os_sorted
```

```python
bb_df = pd.read_csv('branch_and_bound_results.csv')
bb_df
```

```python
# filter only the min values of distance from each unique value in file column
bb_df = bb_df.loc[bb_df.groupby('file')['distance'].idxmin()].reset_index(drop=True)
bb_df
```

```python
# sort using natsort
bb_df['file'] = os_sorted(bb_df['file'])
bb_df
```

```python
# combine all csv in sls folder and open as dataframe
sls_df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "sls/*.csv"))))
sls_df
```

```python
# for each unique value in file column, get the min value of distance
sls_df = sls_df.loc[sls_df.groupby('file')['distance'].idxmin()].reset_index(drop=True)
sls_df
```

```python
# sort using natsort
sls_df['file'] = os_sorted(sls_df['file'])
sls_df
```
