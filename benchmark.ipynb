```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

from tsp.branch_and_bound import BranchAndBound
```

```python
matrix = np.loadtxt('competition/tsp-problem-25-6-100-5-1.txt', skiprows=1)
```

```python
graph = nx.from_numpy_array(matrix)
```

```python
tsp = BranchAndBound(graph)
tsp.solve()
```

```python
def plot_distance(trace):
    """
    Plot the stats of the branch and bound algorithm
  
    Args:
        trace: A list of tuples of the form (time, distance)
    """

    times, distances = zip(*[(t['time'], t['distance']) for t in trace])

    plt.figure(figsize=(10, 6))
    plt.plot(times, distances)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Distance")
    plt.title("Branch and Bound Algorithm Performance")
    plt.show()
```

```python
plot_distance(tsp.trace)
```

```python
def plot_average_branching(trace):
    """
    Plot the stats of the branch and bound algorithm
  
    Args:
        trace: A list of tuples of the form (time, distance)
    """

    times, iterations, total_branching_factor = zip(
        *[(t['time'], t['iterations'], t['total_branching_factor']) for t in trace])
    average_branching_factor = [b / i for b, i in zip(total_branching_factor, iterations)]

    plt.figure(figsize=(10, 6))
    plt.plot(times, average_branching_factor)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Average Branching Factor")
    plt.title("Branch and Bound Algorithm Performance")
    plt.show()
```

```python
plot_average_branching(tsp.trace)
```

```python
def plot_nodes_explored(trace):
    """
    Plot the stats of the branch and bound algorithm
  
    Args:
        trace: A list of tuples of the form (time, distance)
    """

    times, nodes_explored = zip(*[(t['time'], t['nodes_explored']) for t in trace])

    plt.figure(figsize=(10, 6))
    plt.plot(times, nodes_explored)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Nodes Explored")
    plt.title("Branch and Bound Algorithm Performance")
    plt.show()
```

```python
plot_nodes_explored(tsp.trace)
```

```python
def plot_backtracks(trace):
    """
    Plot the stats of the branch and bound algorithm
  
    Args:
        trace: A list of tuples of the form (time, distance)
    """

    times, backtracks = zip(*[(t['time'], t['backtracks']) for t in trace])

    plt.figure(figsize=(10, 6))
    plt.plot(times, backtracks)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Backtracks")
    plt.title("Branch and Bound Algorithm Performance")
    plt.show()
```

```python
plot_backtracks(tsp.trace)
```
