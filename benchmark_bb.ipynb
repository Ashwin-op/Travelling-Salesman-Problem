```python
import pandas as pd
```

```python
df = pd.read_csv('branch_and_bound_results.csv')
df
```

```python
# group by file and find the best distance, then merge back to get the full row
df_best = df.groupby('file').agg({'distance': 'min'}).reset_index()
df_best = df_best.merge(df, on=['file', 'distance'])
df_best
```

```python
# compare to the best known solution
optimal = pd.read_csv('optimal_results.csv')
optimal
```

```python
# take the best known solution based on file column from df_best and create a new column
df_best['optimal'] = df_best['file'].map(optimal.set_index('file')['cost'])
df_best
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# get the error percentage of the best solution
df_best['gap'] = (df_best['distance'] - df_best['optimal']) / df_best['optimal'] * 100
# get the number of nodes based on file name, split by '-' and take the third element
df_best['nodes'] = df_best['file'].str.split('-').str[2]
df_best
```

```python
sns.set_style('whitegrid')
sns.set_context('talk')

sns.scatterplot(data=df_best, x='nodes', y='gap')
plt.xlabel('Nodes')
plt.ylabel('Gap')
plt.title('Optimal Gap')
plt.show()
```

```python
# take the most repeating file column
file = df['file'].value_counts().index[0]
print(file)

# plot the best solution
sns.lineplot(data=df[df['file'] == file], x='time', y='distance')
plt.xlabel('Time')
plt.ylabel('Distance')
# add nodes to the title
plt.title(f'Best Solution for {df[df["file"] == file]["file"].values[0].split("-")[2]} Nodes')
plt.show()
```

```python
# do the above for average_branching_factor
sns.lineplot(data=df[df['file'] == file], x='time', y='average_branching_factor')
plt.xlabel('Time')
plt.ylabel('Average Branching Factor')
plt.title(f'Average Branching Factor for {df[df["file"] == file]["file"].values[0].split("-")[2]} Nodes')
plt.show()
```

```python
# do the above for backtracks
sns.lineplot(data=df[df['file'] == file], x='time', y='backtracks')
plt.xlabel('Time')
plt.ylabel('Backtracks')
plt.title(f'Backtracks for {df[df["file"] == file]["file"].values[0].split("-")[2]} Nodes')
plt.show()
```
